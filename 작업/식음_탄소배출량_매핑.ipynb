```python
import pandas as pd
```

```python
food = pd.read_csv('음식점_좌표변환_대분류_최종.csv')
food.head()
```

```python
food = food[['이름','X','Y','대분류']]
food.head()
```

```python
carbon_factors = {
    '밥류':0.414,
    '국/탕/찌개류':2.736,
    '면류':1.300,
    '생선류':5.933,
    '소고기류':7.833,
    '돼지고기류':1.600,
    '닭고기류':0.300,
    '패스트푸드류':2.600,
    '주류':0.320,
    '탄산음료':0.350,
    '커피':0.180,
    '생수':0.000
}

food['탄소배출량']=food['대분류'].map(carbon_factors)
```

```python
food.to_csv('식음_탄소배출량_매핑.csv')
```
