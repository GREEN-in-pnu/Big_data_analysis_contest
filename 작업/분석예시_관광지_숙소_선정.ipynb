```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2

# Haversine 거리 계산 함수 (단위: km)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return 2 * R * atan2(sqrt(a), sqrt(1 - a))

# 1. 관광지 3곳의 중심점 구하기
center_lat = df['LATITUDE'].mean()
center_lon = df['LONGITUDE'].mean()

# 2. 숙박시설과 중심점 간 거리 계산
숙박['거리_km'] = 숙박.apply(
    lambda row: haversine(center_lat, center_lon, row['위도'], row['경도']),
    axis=1
)

# 3. 가장 가까운 숙박시설 추천 (Top 1)
추천_숙박 = 숙박.sort_values(by='거리_km').head(10)

print("추천 숙박시설:")
print(추천_숙박[['업체명', '위도', '경도', '거리_km']])
# 숙박시설은 T호텔로 지정
# 287        T호텔  35.137569  129.066685  0.938164
```

```python
import osmnx as ox
import networkx as nx
import itertools

# 1. 부산 도로망 로딩
G = ox.graph_from_place("Busan, South Korea", network_type='drive')

# 2. 위치 정보 입력
숙소 = {'name': 'T호텔', 'lat': 35.137569, 'lon': 129.066685}
관광지 = [
    {'name': '민락수변공원', 'lat': 35.155907, 'lon': 129.13435},
    {'name': '다대포해수욕장', 'lat': 35.046382, 'lon': 128.968007},
    {'name': '몰운대', 'lat': 35.046382, 'lon': 128.968007},
]

# 3. 위치별 노드 매핑
def get_node(loc):
    return ox.distance.nearest_nodes(G, loc['lon'], loc['lat'])

숙소['node'] = get_node(숙소)
for g in 관광지:
    g['node'] = get_node(g)

# 4. 순열 생성 및 최적 경로 탐색
min_dist = float('inf')
best_route = []

for perm in itertools.permutations(관광지):
    route_nodes = [숙소] + list(perm) + [숙소]
    total_dist = 0
    path_nodes = []
    
    for i in range(len(route_nodes) - 1):
        try:
            path = nx.shortest_path(G, route_nodes[i]['node'], route_nodes[i+1]['node'], weight='length')
            dist = nx.shortest_path_length(G, route_nodes[i]['node'], route_nodes[i+1]['node'], weight='length')
        except nx.NetworkXNoPath:
            dist = float('inf')
            path = []
        total_dist += dist
        path_nodes += path[:-1]  # 중복 노드 제거용
    
    path_nodes.append(route_nodes[-1]['node'])  # 마지막 노드 추가

    if total_dist < min_dist:
        min_dist = total_dist
        best_route = path_nodes

print(f"최소 이동 거리: {min_dist:.2f} m")

# 5. 지도 시각화
ox.plot_graph_route(G, best_route, route_linewidth=4, node_size=0)
```

```python
import folium
import itertools
from math import radians, sin, cos, sqrt, atan2

# Haversine 거리 계산 함수 (단위: km)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return 2 * R * atan2(sqrt(a), sqrt(1 - a))
# ['민락수변공원', '다대포해수욕장', '몰운대']
# 민락수변공원	35.155907	129.13435
# 다대포해수욕장	35.046382	128.968007	
# 몰운대	35.046382	128.968007
# ▶ 예시 입력
숙소 = {'name': 'T호텔', 'lat': 35.137569, 'lon': 129.066685}
관광지들 = [
    {'name': '민락수변공원', 'lat': 35.155907, 'lon': 129.13435},
    {'name': '다대포해수욕장', 'lat': 35.046382, 'lon': 128.968007},
    {'name': '몰운대', 'lat': 35.046382, 'lon': 128.968007},
]

# ▶ 모든 경로 순열 생성 (관광지 순서만 바뀜)
routes = list(itertools.permutations(관광지들))

# ▶ 최적 경로 탐색
best_route = None
min_distance = float('inf')

for route in routes:
    full_route = [숙소] + list(route) + [숙소]
    total_dist = 0
    for i in range(len(full_route) - 1):
        total_dist += haversine(
            full_route[i]['lat'], full_route[i]['lon'],
            full_route[i+1]['lat'], full_route[i+1]['lon']
        )
    if total_dist < min_distance:
        min_distance = total_dist
        best_route = full_route

# ▶ folium으로 지도 시각화
center_lat = 숙소['lat']
center_lon = 숙소['lon']
m = folium.Map(location=[center_lat, center_lon], zoom_start=14)

# 마커 표시
for place in best_route:
    folium.Marker(
        location=[place['lat'], place['lon']],
        popup=place['name'],
        icon=folium.Icon(color='blue' if place['name'] == '숙소' else 'green')
    ).add_to(m)

# 선 연결
for i in range(len(best_route) - 1):
    start = best_route[i]
    end = best_route[i + 1]
    folium.PolyLine(
        locations=[[start['lat'], start['lon']], [end['lat'], end['lon']]],
        color='red', weight=3
    ).add_to(m)

# ▶ 출력 (Jupyter 또는 HTML로 저장)
m.save('최적관광경로.html')
```
