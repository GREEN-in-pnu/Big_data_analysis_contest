```python
import pandas as pd
```

```python
버스 = pd.read_csv('관광_버스.csv')
지하철 = pd.read_csv('관광_지하철.csv')
숙박 = pd.read_csv('관광_숙박.csv')
식음 = pd.read_csv('관광_식음.csv')
```

```python
지하철
```

```python
관광_3요소 = pd.concat([
    버스[['LOC_NM','LATITUDE','LONGITUDE']], 
    지하철[['관광_지하철']], 
    숙박[['관광_숙박']], 
    식음[['관광_식음']]
], axis=1)
```

```python
관광_3요소.head()
```

```python
관광_3요소.to_csv('관광_3요소.csv')
```
