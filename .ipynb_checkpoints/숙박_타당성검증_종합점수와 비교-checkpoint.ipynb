```python
import pandas as pd
import numpy as np
plt.rcParams['font.family'] = 'Malgun Gothic'
# 데이터 불러오기
숙박2 = pd.read_csv('filtered_data.csv')
명소 = pd.read_csv('관광지_관광명소_정보(좌표가공).csv')

RADIUS_KM_2 = 2.0
nearby_counts = []

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2.0)**2
    c = 2*np.arcsin(np.sqrt(a))
    return R*c

for i, 숙_row in 숙박2.iterrows():
    lat1 = 숙_row['위도']
    lon1 = 숙_row['경도']
    distances = haversine(lat1, lon1, 명소['LATITUDE'].values, 명소['LONGITUDE'].values)
    count = np.sum(distances <= RADIUS_KM_2)
    nearby_counts.append(count)

숙박2['주변명소 수'] = nearby_counts

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
숙박2['숙박점수_norm'] = scaler.fit_transform(숙박2[['숙박점수']])
숙박2['명소수_norm'] = scaler.fit_transform(숙박2[['주변명소 수']])

숙박2['숙박점수_역'] = 1 - 숙박2['숙박점수_norm']
숙박2['종합점수'] = 0.5 * 숙박2['숙박점수_역'] + 0.5 * 숙박2['명소수_norm']

숙박2['주변명소 수'].value_counts()
```

```python
#추가
corr_value = 숙박2[['종합점수', '주변명소 수']].corr().iloc[0, 1]
print(f"종합점수와 주변명소 수의 상관계수: {corr_value:.4f}")
```

```python
#시각화 추가
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x='종합점수', y='주변명소 수', data=숙박2)
plt.title(f"상관계수: {corr_value:.4f}")
plt.show()
```

```python
spearman_corr = 숙박2[['종합점수', '주변명소 수']].corr(method='spearman').iloc[0, 1]
print(f"Spearman 상관계수: {spearman_corr:.4f}")
```

```python
# 🔹 Case 1. 명시적 구간 기반 그룹화 (cut)

import pandas as pd

# 주변명소 수 기반으로 접근성 그룹 나누기
bins = [-1, 0, 5, 15, 숙박2['주변명소 수'].max()]  # 0, 1~5, 6~15, 16+
labels = ['고립', '낮음', '보통', '높음']

숙박2['접근성_그룹'] = pd.cut(숙박2['주변명소 수'], bins=bins, labels=labels)

# 그룹별 통계 확인
print(숙박2['접근성_그룹'].value_counts())
print(숙박2.groupby('접근성_그룹')['종합점수'].describe())

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='접근성_그룹', y='종합점수', data=숙박2)
plt.title("접근성 그룹별 종합점수 분포")
plt.show()
```

```python
# Case 2. k-means 클러스터링

from sklearn.cluster import KMeans
import numpy as np

X = 숙박2[['주변명소 수']].values
kmeans = KMeans(n_clusters=4, random_state=0)
숙박2['접근성_클러스터'] = kmeans.fit_predict(X)

# 클러스터별 중심 확인
print(kmeans.cluster_centers_)

# 시각화
sns.boxplot(x='접근성_클러스터', y='종합점수', data=숙박2)
plt.title("KMeans 클러스터별 종합점수 분포")
plt.show()

# 시각화: 바이올린 플롯
sns.violinplot(x='접근성_클러스터', y='종합점수', data=숙박2)
plt.title("KMeans 클러스터별 종합점수 분포 (Violin Plot)")
plt.show()
```
