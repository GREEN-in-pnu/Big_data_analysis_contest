```python
import pandas as pd
import numpy as np
plt.rcParams['font.family'] = 'Malgun Gothic'
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
ÏàôÎ∞ï2 = pd.read_csv('filtered_data.csv')
Î™ÖÏÜå = pd.read_csv('Í¥ÄÍ¥ëÏßÄ_Í¥ÄÍ¥ëÎ™ÖÏÜå_Ï†ïÎ≥¥(Ï¢åÌëúÍ∞ÄÍ≥µ).csv')

RADIUS_KM_2 = 2.0
nearby_counts = []

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2.0)**2
    c = 2*np.arcsin(np.sqrt(a))
    return R*c

for i, Ïàô_row in ÏàôÎ∞ï2.iterrows():
    lat1 = Ïàô_row['ÏúÑÎèÑ']
    lon1 = Ïàô_row['Í≤ΩÎèÑ']
    distances = haversine(lat1, lon1, Î™ÖÏÜå['LATITUDE'].values, Î™ÖÏÜå['LONGITUDE'].values)
    count = np.sum(distances <= RADIUS_KM_2)
    nearby_counts.append(count)

ÏàôÎ∞ï2['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò'] = nearby_counts

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ÏàôÎ∞ï2['ÏàôÎ∞ïÏ†êÏàò_norm'] = scaler.fit_transform(ÏàôÎ∞ï2[['ÏàôÎ∞ïÏ†êÏàò']])
ÏàôÎ∞ï2['Î™ÖÏÜåÏàò_norm'] = scaler.fit_transform(ÏàôÎ∞ï2[['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò']])

ÏàôÎ∞ï2['ÏàôÎ∞ïÏ†êÏàò_Ïó≠'] = 1 - ÏàôÎ∞ï2['ÏàôÎ∞ïÏ†êÏàò_norm']
ÏàôÎ∞ï2['Ï¢ÖÌï©Ï†êÏàò'] = 0.5 * ÏàôÎ∞ï2['ÏàôÎ∞ïÏ†êÏàò_Ïó≠'] + 0.5 * ÏàôÎ∞ï2['Î™ÖÏÜåÏàò_norm']

ÏàôÎ∞ï2['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò'].value_counts()
```

```python
#Ï∂îÍ∞Ä
corr_value = ÏàôÎ∞ï2[['Ï¢ÖÌï©Ï†êÏàò', 'Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò']].corr().iloc[0, 1]
print(f"Ï¢ÖÌï©Ï†êÏàòÏôÄ Ï£ºÎ≥ÄÎ™ÖÏÜå ÏàòÏùò ÏÉÅÍ¥ÄÍ≥ÑÏàò: {corr_value:.4f}")
```

```python
#ÏãúÍ∞ÅÌôî Ï∂îÍ∞Ä
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x='Ï¢ÖÌï©Ï†êÏàò', y='Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò', data=ÏàôÎ∞ï2)
plt.title(f"ÏÉÅÍ¥ÄÍ≥ÑÏàò: {corr_value:.4f}")
plt.show()
```

```python
spearman_corr = ÏàôÎ∞ï2[['Ï¢ÖÌï©Ï†êÏàò', 'Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò']].corr(method='spearman').iloc[0, 1]
print(f"Spearman ÏÉÅÍ¥ÄÍ≥ÑÏàò: {spearman_corr:.4f}")
```

```python
# üîπ Case 1. Î™ÖÏãúÏ†Å Íµ¨Í∞Ñ Í∏∞Î∞ò Í∑∏Î£πÌôî (cut)

import pandas as pd

# Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò Í∏∞Î∞òÏúºÎ°ú Ï†ëÍ∑ºÏÑ± Í∑∏Î£π ÎÇòÎàÑÍ∏∞
bins = [-1, 0, 5, 15, ÏàôÎ∞ï2['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò'].max()]  # 0, 1~5, 6~15, 16+
labels = ['Í≥†Î¶Ω', 'ÎÇÆÏùå', 'Î≥¥ÌÜµ', 'ÎÜíÏùå']

ÏàôÎ∞ï2['Ï†ëÍ∑ºÏÑ±_Í∑∏Î£π'] = pd.cut(ÏàôÎ∞ï2['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò'], bins=bins, labels=labels)

# Í∑∏Î£πÎ≥Ñ ÌÜµÍ≥Ñ ÌôïÏù∏
print(ÏàôÎ∞ï2['Ï†ëÍ∑ºÏÑ±_Í∑∏Î£π'].value_counts())
print(ÏàôÎ∞ï2.groupby('Ï†ëÍ∑ºÏÑ±_Í∑∏Î£π')['Ï¢ÖÌï©Ï†êÏàò'].describe())

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Ï†ëÍ∑ºÏÑ±_Í∑∏Î£π', y='Ï¢ÖÌï©Ï†êÏàò', data=ÏàôÎ∞ï2)
plt.title("Ï†ëÍ∑ºÏÑ± Í∑∏Î£πÎ≥Ñ Ï¢ÖÌï©Ï†êÏàò Î∂ÑÌè¨")
plt.show()
```

```python
# Case 2. k-means ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ

from sklearn.cluster import KMeans
import numpy as np

X = ÏàôÎ∞ï2[['Ï£ºÎ≥ÄÎ™ÖÏÜå Ïàò']].values
kmeans = KMeans(n_clusters=4, random_state=0)
ÏàôÎ∞ï2['Ï†ëÍ∑ºÏÑ±_ÌÅ¥Îü¨Ïä§ÌÑ∞'] = kmeans.fit_predict(X)

# ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ Ï§ëÏã¨ ÌôïÏù∏
print(kmeans.cluster_centers_)

# ÏãúÍ∞ÅÌôî
sns.boxplot(x='Ï†ëÍ∑ºÏÑ±_ÌÅ¥Îü¨Ïä§ÌÑ∞', y='Ï¢ÖÌï©Ï†êÏàò', data=ÏàôÎ∞ï2)
plt.title("KMeans ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ Ï¢ÖÌï©Ï†êÏàò Î∂ÑÌè¨")
plt.show()

# ÏãúÍ∞ÅÌôî: Î∞îÏù¥Ïò¨Î¶∞ ÌîåÎ°Ø
sns.violinplot(x='Ï†ëÍ∑ºÏÑ±_ÌÅ¥Îü¨Ïä§ÌÑ∞', y='Ï¢ÖÌï©Ï†êÏàò', data=ÏàôÎ∞ï2)
plt.title("KMeans ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ Ï¢ÖÌï©Ï†êÏàò Î∂ÑÌè¨ (Violin Plot)")
plt.show()
```
