```python
import pandas as pd
df = pd.read_excel('부산지역_숙박분야_공공데이터(열린관광시설정보)_20221012.xlsx')
# df.head()
df = df[df['폐업여부'] != 'Y']
df = df[['업체명','카테고리명','경도','위도']]
# df.head()
```

```python
df['카테고리명'].value_counts()
# 글램핑 수가 너무 적음...
```

```python
def classify_accommodation(name):
    name = name.lower()
    if '호텔' in name or '리조트' in name or 'hotel' in name or 'Suite' in name or '레지던스' in name or '브라운도트' in name or '어반스테이' in name or '메리어트' in name or name in ['그랜드조선','팔레드시즈','신라스테이','아난티 힐튼 부산','아난티코브','시그니엘 부산','그랜드엘시티레지던스','엘시티레지던스','일루아','파라다이스','웨스틴 조선 부산']:
        return '호텔/리조트'
    elif '펜션' in name or '민박' in name or '호스텔' in name or '모텔' in name or '하숙' in name or '하우스' in name or 'house' in name or '여관' in name or '장' in name or '무인텔' in name or '여인숙' in name or '숙박' in name or 'guesthouse' in name or '빌라' in name or 'hostel' in name or '탕' in name or '맨션' in name or '콘도' in name:
        return '펜션/민박/유스호스텔'
    elif '캠프' in name or '캠핑' in name or '카라반' in name or '글램핑' in name or '수련관' in name or '연수원' in name or '회관' in name:
        return '캠핑/카라반/글램핑'

df['숙박유형'] = df['업체명'].apply(classify_accommodation)
# df.head()
```

```python
df.loc[df['카테고리명'] == '문화관광/명소', '숙박유형'] = '캠핑/카라반/글램핑'
```

```python
df.head()
```

```python
category_none = df[df['숙박유형'].isna()]
category_none
```

```python
print(category_none.to_string())
```

```python
df = df.dropna(subset=['숙박유형'])
df
```

```python
emission_weights = {
    '호텔/리조트':28.03,
    '펜션/민박/유스호스텔':8.51,
    '캠핑/카라반/글램핑':7.98}

df['숙박점수'] = df['숙박유형'].map(emission_weights)
```

```python
df = df.drop(columns=['카테고리명'])
df
```

```python
df.to_csv('filtered_data.csv', index=False, encoding='utf-8-sig')
```
